```
from pyspark.sql.types import StructType,StructField,IntegerType,StringType,DataType
```

```
race_schema=StructType(fields=[StructField ("raceId",IntegerType(),False),
                                StructField("year",IntegerType(),True),
                                StructField("round",IntegerType(),True),
                                StructField("circuitId",IntegerType(),True),
                                StructField("name",StringType(),True),
                                StructField("date",StringType(),True),
                                StructField("time",StringType(),True),
                                StructField("url",StringType(),True),
                               ])
```

```
races_df=spark.read.option("header",True).schema(race_schema).csv("/FileStore/tables/races.csv")
```

```
from pyspark.sql.functions import current_timestamp,to_timestamp,concat,col,lit
```

```
races_with_timestamp_df=races_df.withColumn("igestion_data",current_timestamp()).withColumn("race_timestamp",to_timestamp(concat(col('date'),lit(' '),col('time')),'yyyy-MM-dd HH:mm:ss'))
```

#####Step-3 Select only the columns required and rename as required

```

races_select_df=races_with_timestamp_df.select(col('raceId').alias('race_id'),col('round'),col('race_timestamp'))
```

```
display(races_select_df)
```

#######Step-4 Write the output to processed container in parquet format

```
races_select_df.write.mode('overwrite').parquet('/FileStore/tables/newraces.csv')
```
